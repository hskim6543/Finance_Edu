```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
import seaborn as sns; sns.set()
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

## BlockChain trend 

```python
kw_list = ["Blockchain"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')
```

```python
df_Blockchain = pytrends.interest_over_time()
```

```python
df_Blockchain.plot()
plt.show()
```

```python
kw_list = ['S&P500']
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')
```

```python
gtrend = pytrends.interest_over_time()
gtrend.index = gtrend.index + pd.Timedelta('1d')   # due to google gives sunday
```

# sp500 index 

```python
sp500 = pd.read_csv("c:/python/data/pff/^GSPC.csv")
data = pd.DataFrame()
data['Date'] = pd.to_datetime(sp500['Date'])
data['Close'] = sp500['Close']
data = data.dropna()
```

```python
data.index = data.Date
data = data.drop('Date', 1)
```

```python
data_w = data.resample('W-MON').last()
equity = data_w[data_w.index >= gtrend.index[0]]
```

```python
print(gtrend.head(), equity.head())
```

```python
#[gtrend['S&P500'], equity['Close']]

df = pd.concat([gtrend['S&P500'], equity['Close']], axis=1)
```

```python
df.head()
```

```python
df['S&P500'].plot(label='google trend')
plt.legend()
df['Close'].plot(label='s&p500 index',secondary_y=True)
plt.legend()
```
